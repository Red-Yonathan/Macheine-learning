```python
df_matrix = df.pivot_table(index=["user_id"],columns=['title'],values=['rating'])
df_matrix
```

```python
# df_matrix = df.apply(lambda x: (x-np.mean(x))/((np.max(x)-np.min(x))),axis =1)
norm = df_matrix.apply( lambda x: (x- np.mean(x))/(np.max(x)-np.min(x)),axis =1 )
```

```python
norm
```

```python
norm.fillna(0,inplace=True)
norm
```

```python
norm_T = norm.T
norm_T
```

```python
norm_T2 = norm_T.loc[:,(norm_T!=0).any(axis=0)]
```
