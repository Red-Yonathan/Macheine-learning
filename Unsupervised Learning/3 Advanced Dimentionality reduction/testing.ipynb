Sure, here are some examples where PCA can be applied while MDS cannot, and vice versa.

PCA can be applied while MDS cannot:
1. Handwritten digit recognition:
   - Scenario: A machine learning system needs to recognize handwritten digits (0-9) from images.
   - Application of PCA: PCA can be used to reduce the dimensionality of the image data by extracting the most important features (principal components) that capture the essential characteristics of the handwritten digits. This can improve the performance of the digit recognition model.
   - Limitation of MDS: MDS is not suitable for this task because it requires a dissimilarity or distance matrix as input, which is not directly available from the image data. MDS is more suitable for analyzing relationships between objects or data points, not for feature extraction or classification tasks.

MDS can be applied while PCA cannot:
2. Mapping perceptual similarities of fragrances:
   - Scenario: A perfume company wants to understand the relationships between different fragrances based on how users perceive their similarities.
   - Application of MDS: MDS can be used to create a visual map of the fragrances based on the perceived similarities or dissimilarities between them, as reported by a group of users. This can help the company identify clusters of fragrances that are perceived as similar, which can inform their product development and marketing strategies.
   - Limitation of PCA: PCA is not suitable for this task because it requires numerical, continuous data, whereas the input for this problem is a dissimilarity matrix based on user perceptions, which are subjective and may not be easily quantified.

In these examples, the nature of the data and the goal of the analysis determine whether PCA or MDS is more appropriate. PCA is better suited for feature extraction and dimensionality reduction tasks, where the input data is numerical and the goal is to capture the maximum variance. MDS, on the other hand, is more suitable for exploring the underlying structure and relationships between data points, especially when the input is a dissimilarity or distance matrix.

The choice between PCA and MDS ultimately depends on the specific problem and the characteristics of the data being analyzed. In some cases, a combination of both techniques may be beneficial, where PCA is used for feature extraction and MDS is used for visualization and exploration of the data structure.
